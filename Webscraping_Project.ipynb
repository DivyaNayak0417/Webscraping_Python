```python
# Importing package for webscraping
from bs4 import BeautifulSoup 
import requests
```

```python
# Fetch html text using requests.get method
html_text = requests.get('https://www.linkedin.com/jobs/search/?currentJobId=3505702084&geoId=103644278&keywords=data%20scientist&location=United%20States&refresh=true').text
```

```python
print(html_text)
```

```python
# Create an object of BeautifulSoup class using lxml as the parser
soup = BeautifulSoup(html_text, 'lxml')
```

```python
soup
```

```python
jobs = soup.find('div', class_='base-search-card__info')
```

```python
print(jobs)
```

```python
# Finding the job_title & repalcing blank spaces
job_title = jobs.find('h3', class_ = 'base-search-card__title').text.replace(' ','')
print(job_title)
```

```python
#Finding location of the job and replacing blank spaces
job_location = jobs.find('span', class_= 'job-search-card__location').text.replace(' ','')
```

```python
# Replacing newline character with no space
job_location.replace('\n', '') 
```
